# [dbo].[v_dispositivo_componentes]

```sql
/****** Object:  View [dbo].[v_dispositivo_componentes]    Script Date: 17-10-2025 18:39:45 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[v_dispositivo_componentes] AS
SELECT
  d.nombre_dispositivo,
  c.componente_id,
  c.nombre_componente,
  c.descripcion_componente,
  c.default_is_recyclable AS is_recyclable,
  c.default_is_reusable   AS is_reusable,
  c.default_is_sellable   AS is_sellable,
  c.default_is_hazardous  AS is_hazardous,
  c.guidance_short,
  c.guidance_url
FROM dbo.dispositivo_componente_mapa m
JOIN dbo.dispositivo_catalogo d ON d.dispositivo_id = m.dispositivo_id
JOIN dbo.componentes_catalogo   c ON c.componente_id = m.componente_id;
GO
```
